### 十一

```python
# Given dataframe tdf with the tweets tokenized, return a list with the k distinct words that are most frequent in the tweets.
def frequent_words(tdf,k):
  words = []
  for i in range(len(tdf)):
    words.extend(tdf['OriginalTweet'][i])
  words_dic = {}
  for word in words:
    if word not in words_dic:
      words_dic[word] = 1
    else:
      words_dic[word] = words_dic[word] + 1
  res = sorted(words_dic.items(),key=lambda x:x[1],reverse=True)
  return [word[0] for word in res[:k]]
```

### 十二

```python
# Given dataframe tdf with the tweets tokenized, remove stop words and words with <=2 characters from each tweet.
# The function should download the list of stop words via:
# https://raw.githubusercontent.com/fozziethebeat/S-Space/master/data/english-stop-words-large.txt
def remove_stop_words(tdf):
  import requests
  r = requests.get('https://raw.githubusercontent.com/fozziethebeat/S-Space/master/data/english-stop-words-large.txt')
  stop_words = r.text.split('\n')
  for i in range(len(tdf)):
    res = []
    for word in tdf['OriginalTweet'][i]:
      if word not in stop_words and len(word)>2:
        res.append(word)
    tdf['OriginalTweet'][i] = res[:]
  return tdf
```

### 十四

```python
# Given a pandas dataframe df with the original coronavirus_tweets.csv data set,
# build a Multinomial Naive Bayes classifier. 
# Return predicted sentiments (e.g. 'Neutral', 'Positive') for the training set
# as a 1d array (numpy.ndarray). 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
def mnb_predict(df):
  target_mapping={'Extremely Negative':0, 'Negative':1, 'Neutral':2,
                'Positive':3, 'Extremely Positive':4}
  # preprocessing the data
  df['SentimentMapped']=df['Sentiment'].map(lambda x:target_mapping[x])
  df['OriginalTweet'] = df['OriginalTweet'].str.lower()
  df['OriginalTweet'] = df['OriginalTweet'].apply(lambda x : re.sub('[^a-zA-Z]', ' ', x))
  df['OriginalTweet'] = df['OriginalTweet'].apply(lambda x : ' '.join(x.split()))
  df['OriginalTweet'] = df['OriginalTweet'].apply(lambda x : x.split())

  train_data = stemming(remove_stop_words(df))

  corpus_train = []
  for i in range(len(train_data)):
    corpus_train.append(' '.join(train_data.OriginalTweet[i]))
  #print(len(corpus_train))

  # convert word to vector with  countvectorizer
  cv = CountVectorizer(max_features=10000)
  X_train_cv = cv.fit_transform(corpus_train).toarray()
  y_train_cv = train_data['SentimentMapped']
  # build a Multinomial Naive Bayes classifier
  mnb_cv=MultinomialNB()
  mnb_cv.fit(X_train_cv, y_train_cv)

  # pred label
  pred_cv = mnb_cv.predict(X_train_cv)

  # mapping to text description
  setiment_mapping={0:'Extremely Negative', 1:'Negative', 2:'Neutral', 
                    3:'Positive', 4:'Extremely Positive'}
  pred_label = np.array([setiment_mapping[x] for x in pred_cv])
  return pred_label
```

```python
# Given a 1d array (numpy.ndarray) y_pred with predicted labels (e.g. 'Neutral', 'Positive') 
# by a classifier and another 1d array y_true with the true labels, 
# return the classification accuracy rounded in the 3rd decimal digit.
def mnb_accuracy(y_pred,y_true):
	num = len(y_pred)
	count = sum([y_pred[i]==y_true[i] for i in range(num)])
	return round(count/num, 3)
```
